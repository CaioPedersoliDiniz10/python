```python
def calcula_ticks(lista,barras):
    menor_valor = lista[0]
    maior_valor = lista[-1]
    intervalo = (maior_valor - menor_valor)/barras
    ticks = [menor_valor]
    ultimo = menor_valor+intervalo
    ticks.append(ultimo)
    for x in range(barras-1):
        ultimo += intervalo
        ticks.append(ultimo)
    return ticks
```
